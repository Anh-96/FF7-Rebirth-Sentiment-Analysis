```python
from helpers import database
import pandas as pd
pd.set_option("display.max_colwidth", None)
```

```python
reviews = database.get_reviews()
reviews.head()
```

```python
def non_alphanumeric_ratio(text):
    if not text:
        return 0.0
    non_alphanumeric = sum(1 for character in text if not character.isalnum() and not character.isspace())
    return non_alphanumeric / len(text)

reviews["non_alphanumeric_ratio"] = reviews["review"].fillna("").astype(str).apply(non_alphanumeric_ratio)
```

```python
ascii_like = reviews[reviews["non_alphanumeric_ratio"] >= 0.8]
ascii_like = ascii_like.sort_values(["votes_up"], ascending=False)[["author_steamid", "votes_up", "non_alphanumeric_ratio", "review"]]
```

```python
def print_ascii_art(review):
    print(f"Author ID: {review['author_steamid']}")
    print(f"Votes: {review['votes_up']}")
    print(f"Non alphanumeric ratio: {review['non_alphanumeric_ratio']}")
    print("=" * 100)
    print(review["review"])
```
